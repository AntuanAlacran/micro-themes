# Redes neuronales (1)

## Ejemplo 1

## Importar librerías necesarias

```python
from keras import Sequential
from keras.layers import Dense
import numpy as np
from matplotlib import pyplot as plt
```

## Generamos los datos
<br>x: Datos de entrada
<br>
y: Datos que deseamos obtener por cada valor x

```python
x = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0])

plt.plot(x,y,'o-b')
plt.plot([4,5.0],[7,9.0],'-b')
plt.grid()
plt.show()
```

## Construimos una mini RNA

```python
# Se crea un modelo secuencial
model = Sequential()

# Se agrega una capa con 1 unidad de salida,
# también se menciona que la entrada es de un valor
model.add(Dense(units=1, input_shape=[1]))

# Se junta todo el modelo, se indica la función de pérdida y el optimizador
model.compile(optimizer='sgd', loss='mean_squared_error')

# Número de Epochs o repeticiones
EPOCHS=500

# Se ajusta el modelo a los valores de x y de y
model.fit(x,y,epochs=EPOCHS)

# Predecimos el valor de salida para x=6
new_val = model.predict([6])
print(new_val)
```

## El modelo ha sido entrenado

Una vez que generamos el modelo, podemos obtener estimaciones para cualquier valor. Lo mismo sucedería en el ejemplo de piedra, papel o tijera: podríamos mandar al modelo cualquier imagen de esas señas para que las pudiera identificar.

```python
xgraph = np.linspace(-10, 10, 21)
ygraph = model.predict(xgraph)

plt.plot(x,y,'o-',markersize=16)
plt.plot(xgraph, ygraph, 's-')
plt.show()
```

## Todo junto para poder modificar las epochs

```python
model = Sequential()
model.add(Dense(units=1, input_shape=[1], activation='linear'))
model.compile(optimizer='sgd', loss='mean_squared_error')

EPOCHS=10

model.fit(x,y,epochs=EPOCHS)

xgraph = np.linspace(-10, 10, 21)
# xgraph = np.linspace(-1, 4, 6)
ygraph = model.predict(xgraph)

plt.plot(x,y,'o-',markersize=12)
plt.plot(xgraph, ygraph, 's-')
plt.grid()
plt.show()

print(y, '\n', model.predict(x).T)

# print(np.abs(y - ygraph.T))
# print(np.mean(np.abs(y - ygraph.T)))
```
